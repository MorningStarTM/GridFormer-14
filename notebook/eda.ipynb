```python
def extract_components(observation):
    # Initialize lists for each component
    gen_list = []
    prod_list = []
    loads_list = []
    lines_or_list = []
    lines_ex_list = []
    line_status_list = []
    topo_vect_list = []
    rho_list = []

    # Helper function to flatten arrays into a list
    def flatten_arrays(array_list):
        return [item for array in array_list for item in array]

    # Extract each component and append to respective lists
    if "gens" in observation:
        gen = observation["gens"]
        gen_list = flatten_arrays([gen["p"], gen["q"], gen["v"]])

    if "prods" in observation:
        prod = observation["prods"]
        prod_list = flatten_arrays([prod["p"], prod["q"], prod["v"]])

    if "loads" in observation:
        loads = observation["loads"]
        loads_list = flatten_arrays([loads["p"], loads["q"], loads["v"]])

    if "lines_or" in observation:
        lines_or = observation["lines_or"]
        lines_or_list = flatten_arrays([lines_or["p"], lines_or["q"], lines_or["v"], lines_or["a"]])

    if "lines_ex" in observation:
        lines_ex = observation["lines_ex"]
        lines_ex_list = flatten_arrays([lines_ex["p"], lines_ex["q"], lines_ex["v"], lines_ex["a"]])

    if "line_status" in observation:
        line_status_list = observation["line_status"].tolist()

    if "topo_vect" in observation:
        topo_vect_list = observation["topo_vect"].tolist()

    if "rho" in observation:
        rho_list = observation["rho"].tolist()

    # Return all lists
    return gen_list, prod_list, loads_list, lines_or_list, lines_ex_list, line_status_list, topo_vect_list, rho_list
```

```python
gen_list, prod_list, loads_list, lines_or_list, lines_ex_list, line_status_list, topo_vect_list, rho_list = extract_components(obs.to_dict())
```

```python
import dill

def load_trajectory(file_path):
    """
    Load a trajectory data file.

    Args:
        file_path (str): Path to the saved trajectory file.

    Returns:
        dict: Loaded trajectory data.
    """
    with open(file_path, 'rb') as f:
        return dill.load(f)
```
